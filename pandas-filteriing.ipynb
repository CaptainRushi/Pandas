**FILTERING**


```python
import pandas as pd
```

```python
df = pd.read_csv("pokemon.csv")
atttack = df[df["Attack"]<50]
print(atttack)
```

```python
spe_att = df[df["Sp. Atk"]>170]
print(spe_att)
```

```python
legend = df[df["Legendary"]==True]
print(legend)
```

```python
water_pok = df[(df["Type 1"]=="Water") | (df["Type 2"]=="Dark")]
print(water_pok
```

```python
fire_pok = df[(df["Type 1"]=="Fire") & (df["Type 2"]=="Flying")]
print(fire_pok)
```
