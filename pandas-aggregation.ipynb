**AGGREGATION**

```python
import pandas as pd
```

** aggregate function = Reduces a set of values into a single summary value
used to summarize and analyze data
Often used with the groupby() function**

```python
df = pd.read_csv("pokemon.csv")
print(df.head())
```

**WHOLE DATAFRAME**

```python
#Whole dataframe
print(df.mean(numeric_only=True))
print(df.sum(numeric_only=True))
print(df.min(numeric_only=True))
print(df.max(numeric_only=True))
print(df.count())
```

**SINGLE COLUMN**

```python
#single column
print(df["Attack"].mean())
print(df["Attack"].sum())
print(df["Attack"].min())
print(df["Attack"].max())
print(df["Attack"].count())
```

```python
print(df)
```

**Group BY**

```python
#group by
group = df.groupby("Type 1")
print(group["Attack"].mean())
print(group["Attack"].count())
print(group["Attack"].min())
print(group["Attack"].sum())
print(group["Attack"].max())
```
