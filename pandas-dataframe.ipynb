**Dataframe = a tabular data structure with rows and columns. (2D) similar to an excel spreadsheet**

```python
import pandas as pd

data  = {
          "name": ["John", "Michael", "Smith"],
          "age": [20, 26, 30]
         }
df = pd.DataFrame(data,index=["employees1","employees2","employees3"])
print(df)
```

```python
print(df.loc["employees1"])
print(df.iloc[2])
```

**Add a new column**

```python
df["job"] = ["accountant","employer","teacher"]
print(df)
```

**add a new row**

```python
new_row = pd.DataFrame([{"name":"john","age":[20],"job":"librarian"}],index=["employees4"])
df = pd.concat([df,new_row])
print(df)
```
